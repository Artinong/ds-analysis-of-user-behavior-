```python
import matplotlib.pyplot as plt
import seaborn as sns
```

in above seaborn provides best representation of data

now countplot for categorial data

```python
for col in df:
    if df[col].dtype == 'O':
        sns.countplot(x=col,data=df)
        plt.show()
```

```python
for col in df:
    if df[col].dtype != 'O':
        sns.histplot(df[col],kde=True)
        plt.show()
```

kernal density estimation (KDE) used to smooth the histogram

```python
sns.boxplot(x='Operating System',y='User Behavior Class',data=df)
```

A box plot helps to maintain the distribution of quantitative data in such a way that it facilitates the comparisons between variables or across levels of a categorical variable.

```python
plt.figure(figsize=(8,8))
sns.lmplot(x='User ID', y='Age', data=df)
plt.show()
```
